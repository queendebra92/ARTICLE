```python
data=pd.read_csv('twelve_year_guardian_league_table.csv')
```

```python
data['Satisfied_with_Course'] = np.where(data['Satisfied_with_Course'] == 0.000000, data['Satisfied_with_Course'].mean(), data['Satisfied_with_Course'])
data['Satisfied_with_Teaching'] = np.where(data['Satisfied_with_Teaching'] == 0.000000, data['Satisfied_with_Teaching'].mean(), data['Satisfied_with_Teaching'])
data['Satisfied_with_Feedback'] = np.where(data['Satisfied_with_Feedback'] == 0.000000, data['Satisfied_with_Feedback'].mean(), data['Satisfied_with_Feedback'])
data['Student_to Staff_Ratio'] = np.where(data['Student_to Staff_Ratio'] == 0.000000, data['Student_to Staff_Ratio'].mean(), data['Student_to Staff_Ratio'])
data['Spend_per_Student'] = np.where(data['Spend_per_Student'] == 0.000000, data['Spend_per_Student'].mean(), data['Spend_per_Student'])
data['Average_Entry_Tariff'] = np.where(data['Average_Entry_Tariff'] == 0.000000, data['Average_Entry_Tariff'].mean(), data['Average_Entry_Tariff'])
data['Value_Added_Score'] = np.where(data['Value_Added_Score'] == 0.000000, data['Value_Added_Score'].mean(), data['Value_Added_Score'])
data['Career_after_15 months'] = np.where(data['Career_after_15 months'] == 0.000000, data['Career_after_15 months'].mean(), data['Career_after_15 months'])
data['Continuation'] = np.where(data['Continuation'] == 0.000000, data['Continuation'].mean(), data['Continuation'])
data['Guardian Score'] = np.where(data['Guardian Score'] == 0.000000, data['Guardian Score'].mean(), data['Guardian Score'])

data.head()
```

# Exploratory graphs

### Data Distribution

```python
#plot the histogram to see the distribution of the point data.
sns.displot(data, x="Guardian Score")
```

### Correlation

```python
corrMatrix = data.corr()
sns.set(rc = {'figure.figsize':(8,8)})
sns.heatmap(corrMatrix, annot = True, cmap= 'coolwarm')
```

### charts comparing variables

```python
sns.histplot(x='Year', data=data);
```

### sorting the data into year

```python
#2021 Guardian Scores
data_2021 =data.loc[data['Year'] == 2021]
data_2021=data_2021.head(18)
sns.set(rc = {'figure.figsize':(20,20)})
#sns.histplot(x='Institution', y ='Guardian Score', data=data_2021);
sns.scatterplot(x='Institution', y ='Guardian Score',  s= 100, palette="deep", data=data_2021);
```

```python
#2020 Guardian Scores
data_2020 =data.loc[data['Year'] == 2020]
data_2020=data_2020.head(18)
sns.set(rc = {'figure.figsize':(20,20)})
#sns.histplot(x='Institution', y ='Guardian Score', data=data_2021);
sns.scatterplot(x='Institution', y ='Guardian Score',  s= 100, palette="deep", data=data_2020);
```
